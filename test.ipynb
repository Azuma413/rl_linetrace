```python
from drqv2.my_simulator import MySimulator2
import matplotlib.pyplot as plt
import cv2
import numpy as np
sim = MySimulator2()
```

```python
obs = sim.obs_map
line_thickness = 1
cv2.line(obs, (250,0), (250,1000), 0, line_thickness)
cv2.line(obs, (500,0), (500,1000), 0, line_thickness)
cv2.line(obs, (750,0), (750,1000), 0, line_thickness)
cv2.line(obs, (0,250), (1000,250), 0, line_thickness)
cv2.line(obs, (0,500), (1000,500), 0, line_thickness)
cv2.line(obs, (0,750), (1000,750), 0, line_thickness)
# y軸を反転
plt.imshow(obs, cmap='gray', origin='lower')
```

```python
obs, reward = sim.simulate(np.array([0.5,0]))
obs = obs.astype(np.float32)
obs1 = obs[:,:,0]
obs2 = obs[:,:,1]
plt.imshow(obs2, cmap='gray', origin='lower')
print(reward)
print("robot position: ", sim.robot_pos)
print("robot average action", sim.action_average)
```
